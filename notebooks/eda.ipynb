```python
def categorize_engine_size(engine_size):
    if engine_size <= 1.5:
        return "Small Car (Compact)"
    elif 1.6 <= engine_size <= 2.5:
        return "Midsize Car"
    elif 2.6 <= engine_size <= 4.0:
        return "SUV / Crossover"
    elif 4.1 <= engine_size <= 6.5:
        return "Sports / Performance Car"
    else:
        return "Heavy-Duty / Truck"

# Apply the function
df['car_category'] = df['engine_size'].apply(categorize_engine_size)

# Check distribution of categories
df['car_category'].value_counts()
```

# Insights


## General Data Distribution Insights
          Fuel consumption (fuel_consumption_comb(l/100km)) and CO₂ emissions are highly skewed—indicating that some vehicles consume significantly more fuel and emit more CO₂ than others.
        Most cars have small to mid-size engines (1.5L - 3.5L), but some outliers (luxury/performance vehicles) have large engines (>5.0L).
        The number of cylinders is mostly 4 or 6, while higher cylinder counts (8-12) are rare and contribute more emissions.


## Correlation Analysis Insights
        CO₂ emissions strongly correlate with fuel consumption (r ≈ 0.85-0.90)

        More fuel consumption = Higher CO₂ emissions.
        Strong indication that fuel efficiency improvements can significantly reduce CO₂ emissions.
        Engine size and CO₂ emissions are highly correlated (r ≈ 0.75)

        Bigger engines generally mean higher emissions, but the fuel type also plays a role.
        Cylinders and CO₂ emissions are moderately correlated (r ≈ 0.60)

        More cylinders often indicate bigger engines, but some cars with advanced technology can have high power output with fewer cylinders.

## Relationship Between Features (Pairplot + Scatterplot Insights)
         Higher engine size = More CO₂ emissions

    Compact cars (~1.5L) emit significantly less CO₂ than large-engine SUVs/trucks (~5.0L).
        Fuel type plays a key role

    Diesel cars tend to have lower fuel consumption but higher CO₂ emissions per unit fuel.
    Premium gasoline vehicles generally have higher emissions than regular gasoline vehicles.
        Some smaller-engine vehicles have surprisingly high emissions
    Could be due to inefficient engines, outdated technology, or lack of hybrid features.

```python
# removing outliear with the IQR methods

for col in df.select_dtypes(include=['int64','float64']):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
        
    IQR=Q3-Q1
        
        #Define the upper and lower bound 
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
        
        
        #removing the outliear 
    df_cleaned=df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
models={
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lassp Regression": Lasso(alpha=0.1),
    "Random Forest" :RandomForestRegressor(n_estimators=100,random_state=42),
    "Gradient Bosstinf":GradientBoostingRegressor(n_estimators=100,learning_rate=0.1,random_state=42),
    "Support Vector Regressor":SVR(kernel='rbf')
    }
results={}

for name,model in models.items():
    model=model.fit(X_train,Y_train)
    y_pred=model.predict(x_test)
    
    results[name]={
        "MAE":mean_absolute_error(y_test,y_pred),
        "MSE":mean_squared_error(y_test,y_pred),
        "R2 Score":r2_score(y_test,y_pred)
        }
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
plt.figure(figsize=(10,5))

#Bar plot for R2 scores
sns.barplot(x=results_df.index,y=results_df["R2 Score"],palette="Set2")
plt.xticks(rotation=15)
plt.ylabel("R2 score")
plt.title("Model comparison(R2 score)")
plt.show()
```

```python
best_model_name=results_df["R2 Score"].idxmax()
best_model=models[best_model_name]

print(f"Best MOdel Selected:{best_model_name}")
      
import joblib
joblib.dump(best_model,"best_co2_emmision_model.pkl")
```
