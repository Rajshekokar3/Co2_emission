```python
q1 = data['co2_emissions'].quantile(0.25)
q3 = data['co2_emissions'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data['co2_emissions'][(data['co2_emissions'] < lower_bound) | (data['co2_emissions'] > upper_bound)] # Access 'co2_emissions' from the DataFrame 'data'
print({len(outliers)})
print(outliers)
print("minimum outlier for co2 emissions column is ",min(outliers))
print("maximum outlier for co2 emissions column is ",max(outliers))
```

# **model building**

# linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score


```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#  Evaluate the Model
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Model Performance:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
#Test the accurcy score
threshold = y_test.mean()
y_pred_class = (y_pred >= threshold).astype(int)
y_test_class = (y_test >= threshold).astype(int)

acc_r = accuracy_score(y_test_class, y_pred_class) * 100
print("Accuracy =", acc_r)
```

# Kmeans Clustering


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)
```

```python
data1["Cluster"] = clusters
```

```python
inertia = kmeans.inertia_
silhouette = silhouette_score(X_scaled, clusters)
davies_bouldin = davies_bouldin_score(X_scaled, clusters)
```

```python
print(f"ðŸ”¹ K-Means Clustering Evaluation for {optimal_k} Clusters:")
print(f"Inertia (WCSS): {inertia:.2f} (Lower is better)")
print(f"Silhouette Score: {silhouette:.2f} (Higher is better, range -1 to 1)")
print(f"Davies-Bouldin Index: {davies_bouldin:.2f} (Lower is better)")
```
